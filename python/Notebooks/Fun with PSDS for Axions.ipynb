```python
from pylab import *
from scipy import fft
from scipy.signal import periodogram as psd
import numpy as np

%matplotlib inline
```

```python
fs = 1000 #Hz, inv. step size (multiply by h gives energy resolution)
dt = 1.0/fs #s, step size
T = 1.0 #s, travel length

for T in [0.9,1,10,100]:
    
    figure(1)
    x=np.arange(0,T,dt)
    y=np.sin(2*pi*x)+np.sin(4*pi*x)+np.sin(5.7*pi*x)+np.random.rand(len(x))*10.0
    plot(x,y,label=T)
    xlim(0,10)
    legend()

    figure(2)
    f,p = psd(y,fs=fs)
    loglog(f,p,label=T)
    legend()

show()

T=100
for sigma in [0.1,1.0,10.0,100.0]:
    
    figure(1)
    x=np.arange(0,T,dt)
    y=np.sin(2*pi*x)+np.sin(4*pi*x)+np.sin(5.7*pi*x)+np.random.rand(len(x))*sigma
    plot(x,y,label=T)
    xlim(0,10)
    legend()

    figure(2)
    f,p = psd(y,fs=fs)
    loglog(f,sqrt(p),label=T,alpha=0.5)
    legend()
    ylim(1e-5,1e1)
    xlim(0.5,10)

show()
```
